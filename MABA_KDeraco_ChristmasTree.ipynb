```python
import datetime

#tree definitions
tree_height = 10
stump = tree_height-1

#xmas countdown
xmas = datetime.date(2021,12,25) - datetime.date.today()

#xmas tree
for row in range(0,tree_height):
    for col_space in range (0,tree_height-row-1):
        print(end=" ") 
    for col_pine in range (0,row+1):
        print("*",end=" ")
    print()

for i in range (stump):
    print(end=" ")
print("#",end=" ")

print()

#xmas countdown
print("There are",xmas, "days left before Christmas")
```
