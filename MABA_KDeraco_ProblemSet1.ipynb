```python
#Problem 1 - Palindromes

import re
def fnc_palindrome(str_input):

    
    return str_input == str_input[::-1] #check if the entered string is equal to reversed string

choice = ''
while choice != 'n':
    
    str_input = input("Please enter a word or sentence to check if palindrome: ")
    str_input=str_input.lower()
    
    str_input2 = re.sub('[^A-Za-z0-9]+', '', str_input) #regex to remove white spaces and symbols
    print ("\nTransformed string: " +str_input2)
    
    #call function to eval and reverse string
    if (fnc_palindrome(str_input2)):
        print("Word or sentence is palindrome.")
        #print(str_input2)
    else:
        print("Word or sentence is not palindrome.")
        #print(str_input2)
    
    choice = input("\nWould you like to try again? [y/n] :")
```
